```python
import os, pandas as pd
import matplotlib.pyplot as plt

print("CWD:", os.getcwd())  # should be path + /OIM_7502_classwork
df = pd.read_csv("../OIM7502_F25/data/wage.csv")
df.head(2)

```

```python
def h(n, title=""):
    print(f"\n{'='*12} Q{n}. {title}")
```

```python
# Q1. Load the dataset  (already done)
h(1, "Load the dataset")
print("Loaded DataFrame with shape:", df.shape)
```

```python
# Q2. First 10 rows
h(2, "Display the first 10 rows")
display(df.head(10))
```

```python
# Q3. Last 10 rows
h(3, "Display the last 10 rows")
display(df.tail(10))
```

```python
# Q4. How many rows
h(4, "How many rows are in the DataFrame?")
print(df.shape[0])
```

```python
# Q5. Metadata
h(5, "Display metadata for the DataFrame")
df.info()
```

```python
# Q6. Data types for two columns (wage, age)
h(6, "Data types for 'wage' and 'age'")
print(df[["wage", "age"]].dtypes)
```

```python
# Q7. Column names
h(7, "Display the column names")
print(df.columns.tolist())
```

```python
# Q8. Index
h(8, "Display the DataFrame’s index")
print(df.index)
```

```python
# Q9. Number of individual elements
h(9, "How many individual elements are in the DataFrame?")
print(df.size)
```

```python
# Q10. How many columns and rows
h(10, "How many columns and rows are there?")
rows, cols = df.shape
print({"rows": rows, "columns": cols})
```

```python
# Q11. Descriptive stats for wage and age
h(11, "Descriptive statistics for wage and age")
display(df[["wage", "age"]].describe())
```

```python

# Q12. Descriptive stats for wage and age by race
h(12, "Descriptive stats for wage and age, by each race")
if "race" in df.columns:
    display(df.groupby("race")[["wage", "age"]].describe())
else:
    print("Column 'race' not found. Available columns:", df.columns.tolist())
```

```python
# Q13. Mean vs median of wage + interpretation
h(13, "Mean and median of wage & interpretation")
mean_w = df["wage"].mean()
median_w = df["wage"].median()
print({"mean_wage": mean_w, "median_wage": median_w})
if mean_w > median_w:
    print("Interpretation: mean > median → right-skewed (a few higher wages pulling mean up).")
elif mean_w < median_w:
    print("Interpretation: mean < median → left-skewed.")
else:
    print("Interpretation: mean ≈ median → roughly symmetric.")
```

```python
# Q14. Display only age and wage columns
h(14, "Display the age and wage columns only")
display(df[["age", "wage"]].head(20))  # remove .head(20) to see all rows
```

```python
# Q15. Are age and wage correlated?
h(15, "Correlation between age and wage")
corr_val = df[["age", "wage"]].corr(numeric_only=True).loc["age", "wage"]
print("Correlation:", corr_val)
```

```python
# Q16. Top 10 salaries (show a few helpful columns if present)
h(16, "Display the top 10 salaries")
keep_cols = [c for c in ["wage","age","education","race","jobclass","occupation","year"] if c in df.columns]
display(df.nlargest(10, "wage")[keep_cols])
```

```python
# Q17. Counts of each race
h(17, "Counts of each race in the dataset")
if "race" in df.columns:
    display(df["race"].value_counts().sort_index())
else:
    print("Column 'race' not found.")
```

```python
# Q18. Proportion of each race
h(18, "Proportion of each race in the dataset (%)")
if "race" in df.columns:
    display((df["race"].value_counts(normalize=True).sort_index()*100).round(2))
else:
    print("Column 'race' not found.")
```

```python
# Q19. Bar plot of job class proportionality
h(19, "Bar plot of job class proportionality")
if "jobclass" in df.columns:
    prop = df["jobclass"].value_counts(normalize=True).sort_index()
    ax = prop.plot(kind="bar")
    ax.set_ylabel("Proportion")
    ax.set_title("Job Class Proportionality")
    plt.show()
else:
    print("Column 'jobclass' not found.")
```
